We want to extract features from the time series. For that we will use simple statistics.


Create a function named calculate_stats_features(x) that calculates some statistical features of a signal x using standard numpy functions: nanpercentile, nanmean, etc.
calculate_stats_features will return a list of features in this order:

0. Max
1. Min
2. Mean
3. Median
4. Variance

```python
def calculate_stats_features(x):
    return [np.nanmax(x), np.nanmin(x), np.nanmean(x), np.nanmedian(x), np.nanvar(x)]
print(calculate_stats_features(arr))
```



Create a function named `calculate_zero_crossing(x)` that calculates the Zero
Crossing of a signal x.

The zero crossing is defined as the number of times the signal changes sign.
For this, you can use the signbit, diff, and nonzero functions from numpy.


```python
def calculate_zero_crossing(x):
    tmp = np.nonzero(np.diff(np.signbit(x)))[0]
    return len(tmp)
print(calculate_zero_crossing(arr))
```

Create a function named **calculate_rms(x)** that returns the Root Mean Square (RMS) of a signal x. We will use the nanmean function instead of the mean function from numpy.

```python
def calculate_rms(x):
    return np.sqrt(np.nanmean(x ** 2))
print(calculate_rms(arr))
```

Create a function named calculate_entropy(x) that calculates the Shannon entropy of a signal x using the entropy function from scipy.stats.

```python
def calculate_entropy(x):
    hist, _ = np.histogram(x, bins=50, density=True)
    probabilities = hist / hist.sum()
    return entropy(probabilities)
print(calculate_entropy(arr))
```

Create a function get_features(x) that combines the features calculated by all previous functions including caculate_stats_features.

```python
def get_features(x):
    res = calculate_stats_features(x)
    res.append(calculate_zero_crossing(x))
    res.append(calculate_rms(x))
    res.append(calculate_entropy(x))
    return res
print(get_features(arr))
```

Load the small ecg dataset
Use your fonction get_features create a new dataframe where you have all the feature as X and y as the label.
Train a random forest on it after doing a train test split if the dataset is not too small

```python
ecg = pd.read_csv('datasets/ecg_small_dataset.csv')
print("Print the dataset")
print(ecg)

label = ecg.iloc[:, 1]
data = ecg.iloc[:, 2:]

print("Print les labels")
print(label)


np_arrays = [row for row in data.to_numpy()]
print("Print les signaux")
print(np_arrays)

x = []
for i in np_arrays:
    x.append(get_features(i))

print("Print les features")
print(x)

y = []
for j in label:
    y.append(j)

print("Print les labels")
print(y)

print("Dataset très petit donc peu de résultats")

x = np.array(x)
y = np.array(y)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Prediction : {y_pred}")
print(f"Valeurs réelles: {y_test}")
print(f"Accuracy: {accuracy:.2f}")
```

We want now to see if a model using only fourier transform could work.

create a function get_fourier_coefficients(ecg)

```python
def get_fourier_coefficients(ecg):
    four = np.fft.rfft(ecg)
    coef = np.abs(four)
    return coef

print(get_fourier_coefficients(x))
```

## Wavelets

We now wants to use another signal decomposition which are called wavelet. Wavelet are a multi-scale function decomposition on a familly of functions generated from what is called a mother wavelet.

Using PyWavelet make a function get_wavelet_coefficients(ecg) that returns the wavelet coefficient of a given ECG

```python
print(type(ecg))
print(ecg.shape)
def get_wavelet_coefficients(ecg):
    ecg = pd.to_numeric(ecg, errors='coerce')
    ecg = ecg.dropna()
    ecg = ecg.to_numpy()
    coef = pywt.wavedec(ecg, wavelet='db4', level=4)
    flat = np.concatenate([c for c in coef])
    return flat
```

Using the get_wavelet_coefficients, create a dataframe when the features are the coefficients and include the label

```python
wavelet_r = []
for _, row in data.iterrows():
    coef = get_wavelet_coefficients(row)
    wavelet_r.append(coef)
wavelet_df = pd.DataFrame(wavelet_r)

#print(wavelet_df)

wavelet_df['label'] = label.reset_index(drop=True)

print(wavelet_df)
```

Train a random forest classifier with such features. DOes the model work

```python
X4 = wavelet_df.iloc[:, :-1]
Y4 = wavelet_df['label']
X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, Y4, test_size=0.3, random_state=42)
model4 = RandomForestClassifier(n_estimators=100, random_state=42)
model4.fit(X_train4, y_train4)
y_pred4 = model4.predict(X_test4)
accuracy4 = accuracy_score(y_test4, y_pred4)
print(f"Accuracy: {accuracy4:.2f}")
```

Add one or several of the previous feature functions and try to train another model

```python
def combine(ecg):
    wave = get_wavelet_coefficients(ecg)
    stat = calculate_stats_features(ecg)
    zer = calculate_zero_crossing(ecg)
    rms = calculate_rms(ecg)
    ent = calculate_entropy(ecg)
    return np.concatenate([wave, stat, [zer, rms, ent]])

ecg = pd.read_csv('datasets/ECG-laurent.csv')
label = ecg.iloc[:, 1]
data = ecg.iloc[:, 2:]

combined = []
for _, row in data.iterrows():
    f = combine(row)
    combined.append(f)
max_len = max(len(f) for f in combined)
padded_features = [np.pad(f, (0, max_len - len(f))) for f in combined]
combined_df = pd.DataFrame(padded_features)
combined_df['label'] = label.reset_index(drop=True)

#print(combined_df)

x = combined_df.iloc[:, :-1]
y = combined_df['label']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model5 = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model5.fit(x_train, y_train)
y_pred = model5.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy GradienBoostingClassifier: {accuracy:.2f}")

model6 = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42, use_label_encoder=False, eval_metric="logloss")
model6.fit(x_train, y_train)
y_pred = model6.predict(x_test)
accuracy2 = accuracy_score(y_test, y_pred)
print(f"Accuracy XGBoost: {accuracy:.2f}")
```

## Deep learning (1D CNN)

Now we want to see if we can skip all theses feature engineering techniques !
Design and train a multi-layer one dimensional CNN using the raw ECG signal as features.


Could you reach or surpass the feature based models ?

```python
A = data.to_numpy()
B = label.to_numpy()

max_len = max([len(row) for row in A])
A = np.array([np.pad(row, (0, max_len - len(row))) for row in A])



enco = LabelEncoder()
B = enco.fit_transform(B)
B = to_categorical(B)

X_train, X_test, y_train, y_test = train_test_split(A, B, test_size=0.3, random_state=42)
print(X_train.shape)
print(X_test.shape)
X_train = X_train[:, :, np.newaxis]
X_test = X_test[:, :, np.newaxis]



model7 = Sequential([
    Conv1D(filters=32, kernel_size=5, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=64, kernel_size=5, activation='relu'),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=256, kernel_size=5, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(y_train.shape[1], activation='softmax')
])

model7.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
h = model7.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), verbose=1)
loss, accuracy = model7.evaluate(X_test, y_test, verbose=0)
print(f"Accuracy of the CNN Model: {accuracy:.2f}")
```
